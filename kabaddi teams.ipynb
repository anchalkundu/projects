```python
import pandas as pd

file_path="C:\\Users\\hp\\Downloads\\teamDetails.csv"
data = pd.read_csv(file_path)
(data.head(12))
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.DataFrame(data)

plt.figure(figsize=(12,6))
sns.barplot(x="TotalPointsScored", y="TeamName",data=df,
            palette="viridis",hue="TeamName",dodge=False,legend=False)
plt.title("Total Points Scored by Teams",fontsize=14)
plt.xlabel("Total Points Scored")
plt.ylabel("Teams")
plt.tight_layout()
plt.show()


numeric_df=df.select_dtypes(include=["number"])
correlation_matrix=numeric_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,fmt=".2f",cmap="coolwarm",cbar=True)
plt.title("Correlation Heatmap of Team Metrics",fontsize=14)
plt.show()
```
