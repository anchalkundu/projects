```python
import pandas as pd


df = pd.read_excel("market data1.xlsx") 
df.head()
```

# Furniture Pricing Across Brand

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.barplot(x="Brand Name", y="Price", data=df, hue="Brand Name", palette="icefire", legend=False, errorbar=('ci',0),width = 0.4)
plt.xticks(rotation=45)
plt.title("Furniture Price Comparison")
plt.xlabel("Brand Name")
plt.ylabel("Price")
plt.show()
```

# Market Strategies Used by Different Brand

```python
plt.figure(figsize=(7, 7))
df["Market Strategy"].value_counts().plot.pie(autopct="%1.1f%%", cmap="Set3")
plt.title("Market Strategy Distribution")
plt.ylabel("")
plt.show()
```

# How Price Affects Ratings

```python
df.columns = df.columns.str.strip()
plt.figure(figsize=(10, 5))
sns.scatterplot(x="Price",y="Customer Rating",hue="Product type",data=df,palette="Set1",s=110,edgecolor="black")

plt.title("Rating By Customers ", fontsize=14)
plt.xlabel("Price (INR)", fontsize=12)
plt.ylabel("Customer Rating", fontsize=12)
plt.legend()
plt.show()
```

# Discount Analysis on Product

```python
df.columns = df.columns.str.strip()

category_discount = df.groupby("Product Category")["discount %"].sum()
plt.figure(figsize=(4,4))
plt.pie(category_discount,labels=category_discount.index,colors=["lightblue", "gold", "violet"],startangle=200,wedgeprops={"edgecolor": "black"})
plt.title("Discount Distribution Across Product Categories", fontsize=14)
plt.show()
```

# Price Trends Across Brands

```python
df.columns = df.columns.str.strip()
brand_price = df.groupby("Brand Name")["Price"].mean().reset_index()
brand_price = brand_price.sort_values(by="Price")

plt.figure(figsize=(10, 5))
sns.lineplot(x="Brand Name",y="Price",data=brand_price,marker="o",linewidth=2,color="b")
plt.xticks(rotation=55,ha="right")
plt.title("Price Variation Across Brands", fontsize=16)
plt.xlabel("Brand Name", fontsize=12)
plt.ylabel("Average Price (INR)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
